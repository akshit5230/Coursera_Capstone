Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import re
import time
from datetime import datetime
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
from urllib.request import urlopen
from bs4 import BeautifulSoup, NavigableString
import requests
from geopy.geocoders import Nominatim
import folium
from IPython.display import display_html
```

Define a function to get scrape table from wikipedia

```python
r = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
content = r.content
soup = BeautifulSoup(content, 'html.parser')
table = soup.table
display_html(str(table),raw=True)
```

Now, scrape the required data from table into a dictionary and create a dataframe

```python
# create a cell as required
dic = { 'PostalCode': [], 'Borough': [], 'Neighborhood': [] }
rows = table.find_all('tr')
    
for row in rows:
    cells = row.find_all('td')      
    for cell in cells:
        all_texts = cell.findAll(text=True)
        dic['PostalCode'].append(all_texts[1].strip())
        dic['Borough'].append(all_texts[2].strip())
        rest_texts = cell.find('span').findAll(text=True)
        if len(rest_texts) > 1:
            neighborhood = ''
            for ttt in rest_texts[1:]:
                neighborhood = neighborhood + ttt
            dic['Neighborhood'].append(neighborhood.strip('()').replace('/', ','))
        else:
            dic['Neighborhood'].append('Not Available')

# create a dataframe using scraped data
df = pd.DataFrame(dic)
print(df.head())
```

Remove cells which dont have borough assigned

```python
index_not_assigned = df[ df['Borough'] == 'Not assigned' ].index
  
df.drop(index_not_assigned, inplace = True)
print(df.head())
```

Since there is no need to further refine the dataframe, 
lets get the coordinates for each postal code

```python
# define a function to get coordinates
def get_coordinates(postal_code):
    lat_lng_coords = None
    geolocator = Nominatim(user_agent="ny_explorer")
    while(lat_lng_coords is None):
        lat_lng_coords = geolocator.geocode('{}, Toronto, Ontario'.format(postal_code))
    latitude = lat_lng_coords.latitude
    longitude = lat_lng_coords.longitude
    return latitude,longitude
```

Get the coordinates for all pincodes

```python
for pin in df['PostalCode']: 
    print('fetching location for ' + pin)
    get_coordinates(pin)  
    print('got location')
```

Finally, lets visualize the coordinates on a map

```python
# create map of Toronto using latitude and longitude values
geolocator = Nominatim(user_agent="ny_explorer")
lat_lng_toronto = geolocator.geocode('Toronto, Ontario')
map_toronto = folium.Map(location=[lat_lng_toronto.latitude, lat_lng_toronto.longitude], zoom_start=10)
    
for i in range(0,len(df)):
    folium.Marker(
      location=[df.iloc[i]['Latitude'], df.iloc[i]['Longitude']],
      popup=df.iloc[i]['Borough'],
    ).add_to(map_toronto)
map_toronto
```
